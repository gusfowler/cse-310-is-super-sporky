```python
n = var('n')
f1(n) = 2^(100*n);
f2(n) = n*sqrt(n);
f3(n) = 2^n;
f4(n) = 2^2^(100*n);

MIN = 1;
MAX = 100;

P = plot(f1(n), (n, MIN, MAX));
P = P + plot(f2(n), (n, MIN, MAX), color='red');
P = P + plot(f3(n), (n, MIN, MAX), color='green');
P = P + plot(f4(n), (n, MIN, MAX), color='yellow');
show(P)
```
